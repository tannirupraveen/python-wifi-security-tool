```python
import os
import subprocess
import re
import requests

def get_connected_devices():
    """
    Fetch a list of devices connected to the Wi-Fi network using ARP.
    """
    try:
        print("\nüì° Fetching connected devices...\n")
        devices_output = subprocess.check_output("arp -a", shell=True).decode()
        devices = re.findall(r"(\d+\.\d+\.\d+\.\d+)\s+([a-fA-F0-9:-]+)", devices_output)

        if devices:
            print("Connected Devices:\n")
            for ip, mac in devices:
                print(f"üîπ IP: {ip} | MAC: {mac}")
        else:
            print("No devices found. Try checking with 'netsh wlan show network mode=bssid'.")

    except Exception as e:
        print("‚ùå Error fetching connected devices:", e)


def change_wifi_password():
    """
    Provide instructions to change the Wi-Fi password via the router's admin panel.
    """
    print("\nüîê Change Wi-Fi Password\n")
    print("1Ô∏è‚É£ Open your browser and go to your router‚Äôs admin panel:")
    print("   - Common addresses: http://192.168.1.1 or http://192.168.0.1")
    print("2Ô∏è‚É£ Log in with your admin username and password.")
    print("3Ô∏è‚É£ Navigate to 'Wireless Settings' > 'Security' > 'Wi-Fi Password'.")
    print("4Ô∏è‚É£ Enter a new password and save changes.")
    print("5Ô∏è‚É£ Restart your router for changes to take effect.\n")


def block_device():
    """
    Block a specific device from your network using its MAC address.
    """
    mac_address = input("Enter MAC address to block (format: 00:1A:2B:3C:4D:5E): ")
    print(f"\nüö´ Blocking device with MAC address: {mac_address}...")

    print("\n‚ùó This must be done via your router's admin panel:")
    print("1Ô∏è‚É£ Log in to your router (http://192.168.1.1 or http://192.168.0.1).")
    print("2Ô∏è‚É£ Go to 'MAC Filtering' or 'Access Control'.")
    print("3Ô∏è‚É£ Add the MAC address and select 'Block'.")
    print("4Ô∏è‚É£ Save and apply settings.\n")


def secure_wifi():
    """
    Provide tips to secure the Wi-Fi network.
    """
    print("\nüîí Wi-Fi Security Tips:\n")
    print("‚úÖ Use a strong Wi-Fi password (min 12 characters).")
    print("‚úÖ Enable WPA3 or WPA2-PSK encryption (avoid WEP).")
    print("‚úÖ Hide your SSID (optional).")
    print("‚úÖ Enable MAC Address Filtering.")
    print("‚úÖ Disable WPS (Wi-Fi Protected Setup) for better security.")
    print("‚úÖ Change the default router admin credentials.")
    print("‚úÖ Regularly check connected devices for unknown users.\n")


def main():
    while True:
        print("\nüì∂ Wi-Fi Security Tool\n")
        print("1Ô∏è‚É£ Show connected devices")
        print("2Ô∏è‚É£ Change Wi-Fi password")
        print("3Ô∏è‚É£ Block a device")
        print("4Ô∏è‚É£ Wi-Fi security tips")
        print("5Ô∏è‚É£ Exit\n")

        choice = input("Select an option (1-5): ")

        if choice == "1":
            get_connected_devices()
        elif choice == "2":
            change_wifi_password()
        elif choice == "3":
            block_device()
        elif choice == "4":
            secure_wifi()
        elif choice == "5":
            print("\nüîÑ Exiting Wi-Fi Security Tool. Stay secure! üîí\n")
            break
        else:
            print("\n‚ùå Invalid choice! Please select a valid option.\n")

if __name__ == "__main__":
    main()
```
